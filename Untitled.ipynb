```python
import folium
from pathlib import Path
# Ваш URL-шаблон
custom_tiles_url = 'http://127.0.0.1:8080/tiles/{z}/{x}/{y}.png'
BASE_DIR = Path(__file__).resolve().parent
# Создаем карту
m = folium.Map(
    location=[43.2432, 76.9],
    tiles=custom_tiles_url,
    attr='© AlatauCityBank | 2025',
    zoom_start=12,
    min_zoom=6,
    max_zoom=18,
    zoomSnap=2,
    zoomDelta=2,
    prefer_canvas=True,  # <<< ВАЖНО: фикс прямоугольника
    width="100%",
    height="720px"
)
```

```python
from branca.element import Element

map_name = m.get_name()

custom_js = f"""
<script>

function initDynamicLayer() {{

    if (typeof {map_name} === 'undefined') {{
        setTimeout(initDynamicLayer, 200);
        return;
    }}

    var MAP = {map_name};
    var pointsLayer = L.layerGroup().addTo(MAP);
    var debounceTimer = null;

    function loadPoints() {{

        var bounds = MAP.getBounds();
        var zoom = MAP.getZoom();

        var min_lat = bounds.getSouth();
        var max_lat = bounds.getNorth();
        var min_lon = bounds.getWest();
        var max_lon = bounds.getEast();

        fetch(`/get_points?min_lat=${{min_lat}}&max_lat=${{max_lat}}&min_lon=${{min_lon}}&max_lon=${{max_lon}}&zoom=${{zoom}}`)
            .then(res => res.json())
            .then(data => {{

                pointsLayer.clearLayers();

                data.forEach(function(p) {{

                    if (p.count !== undefined) {{
                        L.circleMarker([p.lat, p.lon], {{
                            radius: Math.min(25, 6 + Math.log(p.count + 1)),
                            fillOpacity: 0.6
                        }}).addTo(pointsLayer);
                    }} else {{
                        L.circleMarker([p.lat, p.lon], {{
                            radius: 4,
                            fillOpacity: 0.7
                        }}).addTo(pointsLayer);
                    }}

                }});

            }})
            .catch(err => console.log("API error:", err));
    }}

    function delayedLoad() {{
        if (debounceTimer) clearTimeout(debounceTimer);
        debounceTimer = setTimeout(loadPoints, 300);
    }}

    MAP.on("moveend", delayedLoad);
    MAP.on("zoomend", delayedLoad);

    loadPoints();
}}

initDynamicLayer();

</script>
"""

m.get_root().html.add_child(Element(custom_js))
```

```python
MAP_HTML_PATH = BASE_DIR / "map.html"
m.save(MAP_HTML_PATH)
```
